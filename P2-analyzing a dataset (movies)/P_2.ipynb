## Data Cleaning

- Replacing all Na to Unknown for better interpretation of the dataset 
- Casting release_date to date instead of string datatype. 
- Casting budget, revenue, budget_adj, and revenue_adj from float to int datatype for a better representation. 
- Changing all 0 values from budget, revenue, budget_adj, and revenue_adj to prevent distorting results. 
- Remove all unnecessary columns to make the investigation easier.

Convert the following to proper datatypes:

```python
# Convert release_date from string to date: 

df['release_date'] = pd.to_datetime(df['release_date'])

# Convert budget, budget_adj, revenue, and revenue_adj from float to int: 

df['budget'] = df['budget'].astype(int)
df['revenue'] = df['revenue'].astype(int)
df['budget_adj'] = df['budget_adj'].astype(int)
df['revenue_adj'] = df['revenue_adj'].astype(int)
df['release_year'] = df['release_year'].astype(int)
```

In budget, budget_adj, revenue, and revenue_adj replace all 0 values with their according column means: 

```python
# Replace all 0 values in budget, budget, budget_adj, revenue, and revenue_adj with their column means: 

df['budget'] = df['budget'].replace(0,df['budget'].mean())
df['revenue'] = df['revenue'].replace(0,df['revenue'].mean())
df['budget_adj'] = df['budget_adj'].replace(0,df['budget_adj'].mean())
df['revenue_adj'] = df['revenue_adj'].replace(0,df['revenue_adj'].mean())
```

To split contents in columns like production_companies, which uses "|" as a separator, all null cells must be replaced with "Unknown", so our function can work on the correct datatype. 

```python
# Replacing all null cells into "Unknown"
df['genres'] = df['genres'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['production_companies'] = df['production_companies'].fillna('Unknown')
```

Here is a lambda function built to separate "|":

```python
# lambda function: 
df['production_companies'] = df['production_companies'].apply(lambda x: x.split('|')[0])
```

```python
# check the result: 
df.info()
```

To keep things consistent, removing ,imdb_id, cast, homepage, tagline, keywords, and overview from the dataset. 

```python
df = df.drop(['imdb_id', 'cast', 'homepage', 'tagline', 'keywords', 'overview'], axis=1)
```

```python
# Check the result: 

df.info()
```

Now the data has been cleaned.

### Research Question 2  (10 most profited movies over the years)

#### Part 1: Getting profit for all movies

```python
df['profit'] = df['revenue'] - df['budget']
df.head()
```

#### Part 2: Getting profit_adj for all movies

```python
df['profit_adj'] = df['revenue_adj'] - df['budget_adj']
df.head()
```

#### Part 3: highest 10 by profit

```python
df.sort_values(by=['profit'], ascending = False).head(10)
```

The top 10 are: Avatar, Star Wars: The Force Awakens, Titanic, Jurassic World, Furious 7, The Avengers, Harry Potter and the Deathly Hallows: Part 2, Avengers: Age of Ultron, Frozen, and The Net

#### Part 4: sort by profit_adj

```python
df.sort_values(by=['profit_adj'], ascending = False).head(10)
```

The top 10 are: Star Wars, Avatar, Titanic, The Exorcist, Jaws, E.T. the Extra-Terrestrial, Star Wars: The Force Awaken, The Net, One Hundred and One Dalmatians, and The Empire Strikes Back

#### Part 5: Conclusion

We have two lists of the most profited movies over the year. One is adjusted based on the time it was produced while the other is solely based on the amount of profit generated. The two lists are different, yet movies like Star Wars: The Force Awakens and The Net are on both lists, which makes sense because both are within the decade and have high profit in comparison to others. 

#### Part 6: Limitations 

We don't know how the adjusted values are determined, so we cannot compare the adjusted with the unadjusted.
